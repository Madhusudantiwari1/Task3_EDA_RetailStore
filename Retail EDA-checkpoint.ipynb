# Outliers

```python
#Plotting boxplot for Profit
sns.boxplot(x=data['Profit'])
```

```python
#Plotting boxplot for Sales
sns.boxplot(x=data['Sales'])
```

```python
#Plotting boxplot for Quantity
sns.boxplot(x=data['Quantity'])
```

```python
#Plotting boxplot for Discount
sns.boxplot(x=data['Discount'])
```

```python
# finding Interquartile range(IQR)
Q1=data.quantile(0.25)
Q3=data.quantile(0.75)
IQR=Q3-Q1
print(IQR)
```

```python
# Removing Outliers
data=data[~((data<(Q1-1.5*IQR))|(data>(Q3+1.5*IQR))).any(axis=1)]
data.shape
```

```python
data.max()
```

```python
data.min()
```

```python
# finding correlation
c_1 = data["Sales"]
c_2 = data["Discount"]
correl = c_1.corr(c_2)
print(correl)
```

```python
c_1=data["Discount"]
c_2=data["Profit"]
correl=c_2.corr(c_1)
print(correl)
```

```python
c_1 = data["Sales"]
c_2 = data["Quantity"]
correl = c_1.corr(c_2)
print(correl)
```
